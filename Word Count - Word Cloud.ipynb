Importing the python packages required for this word count program

```python
import string                               # this library is collection of string constants and is used for formatting string.
from PIL import Image                       # Image module include functions to load images from files or to create new images.
from wordcloud import WordCloud             # this library is used to give visual representation of text data.
from itertools import islice                # this module is used to implement a number of iterator building blocks.
from nltk.stem import PorterStemmer         # this library is used for stemming process.
from nltk.tokenize import sent_tokenize, word_tokenize
```

#Read the constitu file using file.read().
#Removed the punctuation from the source file (replacing punctuation with the space).
#Lowercase all the words using .lower().

```python
# openig constitu.txt file.
file=open("C:/Users/shubh/Desktop/Python Files/constitu.txt","r")

# reading the constitu.txt file.
mainfile = file.read()              #Read() command is used to read the file 

# Replacing the punctuation from the source file with space
punc = str.maketrans(string.punctuation, ' '*len(string.punctuation))  #str. creates a new string object from the given object

#lowercase words using .lower()  
mainfile = mainfile.translate(punc).lower() 
mainfile
```

Opened the constitution file from the location and kept it in "file" variable. Read the file using .read() command and kept it on "mainfile" variable. Replaced the punctuation from the source file using sting operator. Lower cases the words using lower() command. 

Read stopwords file

```python
# opening stop_words.txt file
stopwords = open("stop_words.txt","r")

# reading the stop_words.txt file
stopwordsfile = stopwords.read()

# checking source file words with stopwords file, using loop.
wordlist=mainfile.split()   

filt_data=[s for s in wordlist if not s in stopwordsfile]
filt_data
```

Opened and read the stopwords file and kept it in "stopwordsfile" variable. Checked the source file words with stopwords file and filtered the file and kept it in filt_data.

#Sorting in order of frequency of occurrence of the words found with the highest occurrences coming first.

```python
# created a dictionary "dic" and counting number of times each word comes up in list of words (in dictionary)
dic={}
for word in filt_data:
    dic[word]=dic.get(word,0)+1
dic

# sorting with highest count    
import operator
sorted_file = sorted(dic.items(), key=operator.itemgetter(1),reverse=True)
sorted_file
```

Created a dictionary "dic", counted the frequency of words occuring in filt_data file and stored in dic. Sorted the words in ascending order using sorted operator and stored in "sorted_file" and displayed the words sorted in order of frequecy of occurance.

# Displaying top 20 words

```python
topwords=dict(sorted_file)

#list the words and stored in top20 varible
top20 = list(islice(topwords.items(),20))
top20
```

Displayed the list of top 20 words according to the highest frequency occurance using list command.
